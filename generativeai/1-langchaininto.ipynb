### Streaming and Batch

### Streaming

```python
model.invoke("Write me a 500 words paragraph on AI.")
```

```python
for chunk in model.stream("Write me a 1000 words paragraph on AI."):
    print(chunk.text, end="|", flush=True)
```

```python
model.invoke("Why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers ?"):
    print(chunk.text, end="|", flush=True)
```

### Batching

```python
model = init_chat_model("google_genai:gemini-2.5-flash")
```

```python
responses = model.batch(
    [
    "Why do parrots have colorful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"
    ]
)
for response in responses:
    print(response)
```

```python
responses = model.batch(
    [
    "Why do parrots have colorful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"
    ],
    config={
        'max_concurrency':5, #Limit to 5 parallel cells 
    }
)
for response in responses:
    print(response)
```
